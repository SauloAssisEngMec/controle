## Exemplo utilizando Espaço de Estados

Trata-se de problema clássico do sistema massa-mola, ver slide 11-13 do arquivo: "2a_Controle - Modelos DORF"


### Traduzido do original:

https://python-control.readthedocs.io/en/latest/secord-matlab.html

```python
# secord.py - demonstrate some standard MATLAB commands
# RMM, 25 May 09

# Parâmetros do sistema
m = 250.0           # smassa
k = 40.0            # cosntante da mola
b = 60.0            # constante de amortecimento

# System matrices - abrindo como um espaço de estados
A = [[0, 1.], [-k/m, -b/m]]
B = [[0], [1/m]]
C = [[1., 0]]
sys = ss(A, B, C, 0)
```

```python
# Resposta ao degrau
plt.figure(1)
yout, T = step(sys)
plt.plot(T.T, yout.T)
plt.show(block=False)
```

### Tarefa 1

Procure variar os valores de k/m e b/m para reproduzir os gráficos do slide 13.

```python
# Plotando o Lugar das raízes do sistema
rlocus(sys)
plt.show()
```

## Plotando o diagrama de Bode e de Nyquist

```python
plt.figure(2)
mag, phase, om = bode(sys, logspace(-2, 2), plot=True)
plt.show(block=False)
```

```python
plt.figure(3)
nyquist(sys, logspace(-2, 2))
plt.show(block=False)
```
