Este notebook demonstra a estimaÃ§Ã£o de parÃ¢metros de um motor CC acionando uma carga mecÃ¢nica constante e um controlador PD, conforme descrito em: https://github.com/dbr-ufs/controle/blob/main/system_identification-final.pdf 



onde: $\omega$ = Velocidade do motor, 
$\it{V}$ = TensÃ£o aplicada na armadura,
$\it{Ia}$ = corrente de armadura,
$\it{L}$ = IndutÃ¢ncia da armadura,
$\it{e_a}$ = Forca contra eletromotriz,
$\it{R}$ = ResistÃªncia do enrolamento da armadura,
$\it{T}$ = Torque interno do motor = $k ia$,
$\it{J}$ = Momento de inercia da carga,
$\it{f}$ = Atrito viscoso no eixo.

Tem-se desta forma as duas equaÃ§Ãµes que descrevem o sistema para a parte elÃ©trica
e mecÃ¢nica, sendo elas:

\begin{equation}
V(t)-R\cdot Ia(t) - L\frac{\mathrm{d} Ia(t)}{\mathrm{d} t}-e_a(t) = 0
\label{eq:motor_elet}
\end{equation}

\begin{equation}
J \dot{\omega}(t)= T-f\omega(t)
\label{eq:motor_mec}
\end{equation}

A conversÃ£o eletromecÃ¢nica ocorre na armadura, onde se iguala a potÃªncia interna desenvolvida pelo motor:

\begin{equation}
P_{int} = e_a Ia = T \omega(t)
\end{equation}

A partir destas equaÃ§Ãµes diferenciais pode-se chegar ao sistema de equaÃ§Ãµes de estado 
descrito em:

\begin{equation}
\begin{matrix}
\begin{bmatrix}
\dot{Ia}\\ 
\dot{\omega} \\
\dot{\theta}
\end{bmatrix}= &
\begin{bmatrix}
-R/L & -k/L & 0\\ 
k/J & -f/J & 0 \\
0 & 1 & 0
\end{bmatrix}&
\begin{bmatrix}
Ia\\ 
\omega \\
\theta
\end{bmatrix} + &
\begin{bmatrix}
1\\ 
0\\
0
\end{bmatrix} V
\end{matrix}
\label{eq:motor_eq_estados}
\end{equation}

Com esse sistema de equaÃ§Ãµes de estado, Ã© possÃ­vel obter-se as funÃ§Ãµes de transferÃªncias,
que descrevem o sistema, para uma entrada e uma saÃ­da. Encontra-se, portanto a funÃ§Ã£o de transferÃªncia do motor, que relaciona a posiÃ§Ã£o angular com a tensÃ£o de entrada do motor:

\begin{equation}
G(s)=\frac{k/J_L}{s(s+f/L)(s+R/L)+s\;k^2/J_L}
\label{eq:motor_pos_tf}
\end{equation}

Devido a dinÃ¢mica do sistema mecÃ¢nico ser muito mais lenta que a dinÃ¢mica do sistema elÃ©trico, 
pode-se desconsiderar a influencia do polo elÃ©trico do sistema, ficando a funÃ§Ã£o de transferÃªncia
assim:

\begin{equation}
G(s)=\frac{\frac{k}{J_L}}{s(s+\frac{f_L+k_1 k_2}{J_L})}
\nonumber
\end{equation}

A equaÃ§Ã£o acima representa o comportamento da posiÃ§Ã£o angular do motor baseado na entrada de tensÃ£o aplicada sobre seus terminais.

Para uma entrada em rampa, a coleta de dados foi realizada utilizando o seguinte controlador PD:

$PD(s) = k_p + k_d.s$.

Fechando a malha da realimentaÃ§Ã£o unitÃ¡ria, observe que a funÃ§Ã£o de transferÃªncia de malha fechada tem o seguinte formato: 

$T(s) = \frac{T_z.s+1}{T_w^2.s^2+2.\zeta.T_w.s+1}$, 

onde $T_z = \frac{k_d}{k_p}$ (devido ao controlador PD). Cabe a nÃ³s, portanto, identificar os demais parÃ¢metros $\zeta$ e $T_w$.

## Passo 3. Definir a funÃ§Ã£o objetivo

Para isso, simplesmente foi escolhido a diferÃªncia entre o valor estimado e o valor medido. Aplicando a funÃ§Ã£o control.forced_response(sys,t,u) que faz a convoluÃ§Ã£o da funÃ§Ã£o de transferÃªncia com a entrada, encontra-se $y_e$ (valor estimado de y).

```python
kp = 3      # ganho proporcional
kd = 0.05   # ganho derivativo
Tz = kd/kp

def residuals(parameters, df):
    Zeta, Tw = parameters
    t, y, u = df['t'], df['y'], df['u']
    sys = tf([Tz,1],[Tw**2,2*Zeta*Tw,1])
    t, y_e = control.forced_response(sys,t,u)
    return y_e-y
```

Para ilustrar, fazendo um chute inicial dos dois parÃ¢metros a ser estimados, avalia-se o resÃ­duo no tempo.

```python
parameter_names = ['Zeta', 'Tw']
parameter_guess = [1, 0.025]

r = residuals(parameter_guess, df)
r.plot()
```

## Passo 3.1. Avaliando o "mÃ©rito" dos parÃ¢metros estimados

Existem diversas formas de se avaliar o mÃ©rito, uma das mais usuais Ã© o erro quadrÃ¡tico, sendo ele:

\begin{align*}
SOS & = \sum_k r_k^2
\end{align*}

O objetivo serÃ¡ portanto, encontrar os parÃ¢metros que minimizem a soma dos erros quadrÃ¡ticos.

Definindo essas funÃ§Ãµes, observe que `sos` calcula a soma de todos os resÃ­duos e eleva ao quadrado, jÃ¡ `best_fit` chama o algoritmo de minimizaÃ§Ã£o de Nelder-Mead implementado na biblioteca: `scipy.optimize.fmin`.

ReferÃªncia:
Nelder, J.A. and Mead, R. (1965), âA simplex method for function minimizationâ, The Computer Journal, 7, pp. 308-313

```python
def sos(parameters, df):
    return sum(r**2 for r in residuals(parameters, df))

def best_fit(fcn, df, disp=1):
    return optimize.fmin(fcn, parameter_guess, args=(df,), disp=disp)

parameter_fit = best_fit(sos, df)

for name,value in zip(parameter_names, parameter_fit):
    print(name, " = ", round(value,4))
```

## Passo 4. Avaliar a qualidade do ajuste

```python
t, y, u = df['t'], df['y'], df['u']
Zeta,Tw = parameter_fit
sys = tf([Tz,1],[Tw**2,2*Zeta*Tw,1])
t1, y_e = control.forced_response(sys,t,u)
r = y_e-y
r.plot()
```
